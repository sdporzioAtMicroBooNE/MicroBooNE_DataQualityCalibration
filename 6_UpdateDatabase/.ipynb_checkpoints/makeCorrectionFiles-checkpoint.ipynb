## Time corrections

```python
### Load data
inFile = 'Test/second_global_dqdx_all_planes.txt'
lines = open(inFile).readlines()

time, p0, p1, p2 = [], [], [], []
for line in lines:
    words = line.split(' ')
    if len(words)==9:
        t = words[0]
        day = int(t.split('/')[1])
        month = int(t.split('/')[0])
        if month < 6: year = 2018
        else: year = 2017
        t = dt(year,month,day,0,0,0)
        time.append(t)
        p0.append(float(words[2]))
        p1.append(float(words[4]))
        p2.append(float(words[6]))
        
df = pd.DataFrame(OD([('time',time),('p0',p0),('p1',p1),('p2',p2)]))
df[0:2]

## Find bad days from plane 2
i = df[df['p2']>208].index.values
i_bad = df[df['p2']<=208].index.values

mean0 = float(df['p0'][i].mean())
mean1 = float(df['p1'][i].mean())
mean2 = float(df['p2'][i].mean())

dfCorr = df.copy()
dfCorr['p0'] = mean0/df['p0']
dfCorr['p1'] = mean1/df['p1']
dfCorr['p2'] = mean2/df['p2']

dfCorr[0:2]
```

```python
### Make plots to make sure you're doing it right
figsize = (16,9)

fig = plt.figure(figsize=figsize)
plt.subplot(3,2,1)
plt.plot(df['time'][i],df['p0'][i],ls='',color='C0',marker='o')
plt.plot(df['time'][i_bad],df['p0'][i_bad],ls='',color='red',marker='o')
plt.axhline(mean0,ls='--',color='C0')
# plt.ylim(120,220)
plt.subplot(3,2,3)
plt.plot(df['time'][i],df['p1'][i],ls='',color='C1',marker='o')
plt.plot(df['time'][i_bad],df['p1'][i_bad],ls='',color='red',marker='o')
plt.axhline(mean1,ls='--',color='C1')
# plt.ylim(120,220)
plt.subplot(3,2,5)
plt.plot(df['time'][i],df['p2'][i],ls='',color='C2',marker='o')
plt.plot(df['time'][i_bad],df['p2'][i_bad],ls='',color='red',marker='o')
plt.axhline(mean2,ls='--',color='C2')
# plt.ylim(120,220)

plt.subplot(3,2,2)
plt.plot(dfCorr['time'][i],dfCorr['p0'][i],ls='',color='C0',marker='o')
plt.plot(dfCorr['time'][i_bad],dfCorr['p0'][i_bad],ls='',color='red',marker='o')
plt.axhline(1.,ls='--',color='C0')
plt.ylim(0.9,1.1)
plt.subplot(3,2,4)
plt.plot(dfCorr['time'][i],dfCorr['p1'][i],ls='',color='C1',marker='o')
plt.plot(dfCorr['time'][i_bad],dfCorr['p1'][i_bad],ls='',color='red',marker='o')
plt.axhline(1.,ls='--',color='C1')
plt.ylim(0.9,1.1)
plt.subplot(3,2,6)
plt.plot(dfCorr['time'][i],dfCorr['p2'][i],ls='',color='C2',marker='o')
plt.plot(dfCorr['time'][i_bad],dfCorr['p2'][i_bad],ls='',color='red',marker='o')
plt.axhline(1.,ls='--',color='C2')
plt.ylim(0.9,1.1)
plt.show()
```

```python
### Save to file
epoch = datetime.datetime(1970,1,1)

pb = f.pbar(len(dfCorr))
for i,row in dfCorr.iterrows():
    
    time = row['time']
    timestamp = (time-epoch).total_seconds()

    name_out = 'Output/dqdx_t_data/T_correction_factors_%i_v1r3.dat' %(timestamp)
    with open(name_out,'wb') as fileOut:
        fileOut.write('# time %i\n' %timestamp)
        fileOut.write('# channel correction correction_err\n')
        fileOut.write('0 %f %f\n' %(row['p0'],0.))
        fileOut.write('1 %f %f\n' %(row['p1'],0.))
        fileOut.write('2 %f %f\n' %(row['p2'],0.))
    pb.update()
```
